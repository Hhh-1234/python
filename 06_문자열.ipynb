# 📘 파이썬 기초: 06. 문자열


## 1. 문자열의 개요
* **문자열의 개념**: 0개 이상의 문자들이 모인 집합으로, 작은따옴표(`'`)나 큰따옴표(`"`)로 감싸 표현합니다.
* **문자열 연결하기**: `+` 연산자를 사용하여 두 문자열을 하나로 합칠 수 있습니다.
* **문자열 반복하기**: `*` 연산자를 사용하여 문자열을 지정한 횟수만큼 반복할 수 있습니다.

## 2. 문자열 다루기
* **인덱싱(Indexing)**: 문자열의 특정 위치에 있는 문자에 접근하는 방법입니다. 인덱스는 **0번**부터 시작합니다.
* **슬라이싱(Slicing)**: 문자열의 일부분을 추출하는 방법입니다. `[시작:끝]` 형식에서 끝 인덱스 직전까지 추출합니다.
* **문자열 함수 및 메서드**:
    * `len()`: 문자열의 길이를 반환합니다.
    * `upper()` / `lower()`: 대문자/소문자로 변환합니다.
    * `replace()`: 특정 문자를 다른 문자로 치환합니다.
    * `split()`: 특정 구분자를 기준으로 문자열을 분리하여 리스트로 만듭니다.
    * `count()`: 특정 문자의 출현 횟수를 셉니다.
    * `find()` / `index()`: 특정 문자의 위치를 탐색합니다.
