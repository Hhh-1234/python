```python
import requests
url = "https://www.omdbapi.com/"
params = {
    "apikey": "9f67b856",
    "s" : "Avengers",# t 대신 s 사용 - 제목에 avengers 포함된 영화 전부 검색
    "type" : "movie"
}

response = requests.get(url, params=params)
data = response.json()
data
```

```python
#검색결과 리스트 출력
movies = data["Search"] # 'search' 대신 'Search' 사용, 변수명도 'movies'로 변경
for movie in movies:
  print(movie['Title'], movie['Year']) # 괄호 오류 수정 및 'title', 'year' 대신 'Title', 'Year' 사용
```

```python
import pandas as pd

if 'Search' in data: # 'search' 대신 'Search' 사용
  m_df = pd.DataFrame(data['Search'])
else:
  m_df = pd.DataFrame([data]) # 'Search' 키가 없으면, data자체가 단일 객체이므로 리스트로 감싸서 dataframe으로 만듭니다.

display(m_df.head()) # .head 뒤에 괄호 추가
```

```python
m_df = m_df.iloc[0:4]# 인덱스 0부터 3까지의 행만 선택합니다.
display(m_df)
```

```python
import requests
code = "005930"
url = "https://finance.naver.com/item/main/naver?code={}"

response = requests.get(url.format(code))
response.encoding = "euc-kr" # 담당 텍스트의 인코딩을 'euc-kr'로 변경
print(response.text)
```

```python
from bs4 import BeautifulSoup # html 및 xml 파일을 피싱

soup = BeautifulSoup(response.text, "html.parser") #응답 HTML을 파싱하여 soup 객체 새서
```

```python
#현재가 태그 찾기
price_tag = soup.find('p', attrs={'class' : 'no_today'})
current_price = price_tag.find('span', attrs={'class' : 'blind'}).get_text()
print(f"현재가: {current_price}")
```
